```python
import pandas as pd
import json


df = pd.DataFrame()

for js in zf.namelist():
    if js.endswith('.json'):
        try:
            json_file = json.load(zf.open(js))

            data = {
                'id': json_file.get('name'),
                'name_id': json_file.get('id'),
                'lcd': json_file.get('lcd'),
                'pld': json_file.get('pld'),
                'void_fraction': json_file.get('void_fraction'),
                'surface_area_m2g': json_file.get('surface_area_m2g'),
                'surface_area_m2cm3': json_file.get('surface_area_m2cm3'),
                'pressure_0.01(bar)': json_file.get('isotherms', [])[0]['isotherm_data'][0]['total_adsorption'] if json_file.get('isotherms') else None
            }

            # Create a temporary DataFrame for the current 'data'
            temp_df = pd.DataFrame([data])

            # Concatenate the temporary DataFrame with the main DataFrame
            df = pd.concat([df, temp_df], ignore_index=True)

        except json.JSONDecodeError:
            print(f"Skipping {js} as it is not a valid JSON file.")


```
